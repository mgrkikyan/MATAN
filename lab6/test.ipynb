```python
import numpy as np
import matplotlib.pyplot as plt

#  
def f(x, y):
    return np.sin(2*x) * np.cos(3*y - 1)

#  
def df_dx(x, y):
    return 2 * np.cos(2*x) * np.cos(3*y - 1)

def df_dy(x, y):
    return -3 * np.sin(2*x) * np.sin(3*y - 1)

#      
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

#      
z_min = Z.min()
z_max = Z.max()

#   levels
levels = np.linspace(z_min, z_max, 10)

#   
plt.contour(X, Y, Z, levels=levels)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')

#   
A = (0.9, 0)
B = (1, 1)
C = (-1.1, 0)
D = (0, -1.3)

plt.plot(A[0], A[1], 'ro', label='A')
plt.plot(B[0], B[1], 'go', label='B')
plt.plot(C[0], C[1], 'bo', label='C')
plt.plot(D[0], D[1], 'yo', label='D')

#    
grad_A = np.array([df_dx(A[0], A[1]), df_dy(A[0], A[1])])
grad_B = np.array([df_dx(B[0], B[1]), df_dy(B[0], B[1])])
grad_C = np.array([df_dx(C[0], C[1]), df_dy(C[0], C[1])])
grad_D = np.array([df_dx(D[0], D[1]), df_dy(D[0], D[1])])

#    
scale = 0.5

plt.quiver(A[0], A[1], grad_A[0]*scale, grad_A[1]*scale, color='r')
plt.quiver(B[0], B[1], grad_B[0]*scale, grad_B[1]*scale, color='g')
plt.quiver(C[0], C[1], grad_C[0]*scale, grad_C[1]*scale, color='b')
plt.quiver(D[0], D[1], grad_D[0]*scale, grad_D[1]*scale, color='y')

plt.legend()
plt.show()
```
